# Motivational Example


## Artificial Neural Networks (ANN)

```python
import numpy as np

# Install TensorFlow
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass

import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0


x_train_dat = x_train[:50000]
x_val = x_train[50000:]
y_train_dat = y_train[:50000]
y_val = y_train[50000:]
```

```python
ann = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(units=512, activation='relu'),
    tf.keras.layers.Dense(units=10,activation='softmax'),
])
ann.summary()
ann.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
ann.fit(x_train_dat, y_train_dat, epochs=10,
        batch_size=64,
        validation_data=(x_val, y_val))
eval_loss, eval_acc = ann.evaluate(x_test, 
                                   y_test)
print(eval_acc)
```

## Convolutional Neural Networks (CNN)

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape((60000, 28, 28, 1))
x_test = x_test.reshape((10000, 28, 28, 1))
x_train, x_test = x_train.astype('float32')  / 255.0, x_test.astype('float32')  / 255.0

x_train_dat = x_train[:50000]
x_val = x_train[50000:]
y_train_dat = y_train[:50000]
y_val = y_train[50000:]
```

```python
cnn = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dense(10, activation='softmax')
])

cnn.summary()

cnn.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
cnn.fit(x_train_dat, y_train_dat, epochs=10,
        batch_size=64,
        validation_data=(x_val, y_val))
eval_loss, eval_acc = cnn.evaluate(x_test, y_test)
print(eval_acc)
```

### Without padding

```python
cnn = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)), 
    tf.keras.layers.MaxPooling2D(2, 2), 
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), 
    tf.keras.layers.MaxPooling2D(2, 2), 
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), 
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dense(10, activation='softmax')
])
cnn.compile(optimizer='rmsprop',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
cnn.fit(x_train_dat, y_train_dat, epochs=10,
        batch_size=64, 
        validation_data=(x_val, y_val));
print(cnn.summary())
test_loss, test_acc = cnn.evaluate(x_test, y_test)
print(test_acc)
```

### With padding="same"

```python
cnn = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2, 2), 
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), 
    tf.keras.layers.MaxPooling2D(2, 2), 
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'), 
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dense(10, activation='softmax')   
])
cnn.compile(optimizer='rmsprop',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
cnn.fit(x_train_dat, y_train_dat, epochs=10,
        batch_size=64, 
        validation_data=(x_val, y_val));
print(cnn.summary())
test_loss, test_acc = cnn.evaluate(x_test, y_test)
print(test_acc)
```
